```python
sns.set_style("darkgrid")
matplotlib.rcParams['figure.figsize'] = (9, 7)
```

```python
a1=df.NormalMM
a1
a2=df.ActualMM
a3=df.Location
```

```python
plt.plot(df.Location,df.NormalMM,linewidth=6,linestyle = '-')
plt.plot(df.Location,df.ActualMM,'r--o',linewidth=4)
plt.title('RainFall Data in mm \n from 11-1 to 11-8')

plt.legend(['Normal', 'Actual'])
plt.xticks(rotation = 90.5)
plt.show()
```

```python
# Create a scatter plot
sns.scatterplot(a1,a2,hue=a3,
                s=100);

# Chart title
plt.title("RainFAll data ");
```

```python
plt.title("RainFall Data")

sns.distplot(a1, kde=False);

sns.distplot(a2, kde=False);
```

```python
plt.title("RainFall Data")

sns.kdeplot(a1, a2, shade=True, shade_lowest=False);
```

```python
# Chart title
plt.title("Daily Total Bill")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(a2, a1);
```
